```python
import numpy as np
import time
import copy

# Função para gerar uma matriz aleatória
def generate_random_matrix(m=3, n=3, min_limit=0, max_limit=10):
    return np.random.randint(min_limit, max_limit + 1, size=(m, n))

# Inputs do usuário para definir a matriz
user_m = int(input("Número de linhas (m): "))
user_n = int(input("Número de colunas (n): "))
user_min_limit = int(input("Valor mínimo: "))
user_max_limit = int(input("Valor máximo: "))

# Gerar a matriz baseada nos inputs do usuário
matrix = generate_random_matrix(user_m, user_n, user_min_limit, user_max_limit)
print("Matriz original:")
print(matrix)

# Função de Selection Sort
def selection_sort(matrix):
    sorted_matrix = copy.deepcopy(matrix)
    for row in sorted_matrix:
        for i in range(len(row)):
            min_idx = i
            for j in range(i + 1, len(row)):
                if row[j] < row[min_idx]:
                    min_idx = j
            row[i], row[min_idx] = row[min_idx], row[i]
    return sorted_matrix

# Função de Insertion Sort
def insertion_sort(matrix):
    sorted_matrix = copy.deepcopy(matrix)
    for row in sorted_matrix:
        for i in range(1, len(row)):
            key = row[i]
            j = i - 1
            while j >= 0 and row[j] > key:
                row[j + 1] = row[j]
                j -= 1
            row[j + 1] = key
    return sorted_matrix

# Função de Bubble Sort
def bubble_sort(matrix):
    sorted_matrix = copy.deepcopy(matrix)
    for row in sorted_matrix:
        n = len(row)
        for i in range(n):
            for j in range(0, n - i - 1):
                if row[j] > row[j + 1]:
                    row[j], row[j + 1] = row[j + 1], row[j]
    return sorted_matrix

# Função de Shell Sort
def shell_sort(matrix):
    sorted_matrix = copy.deepcopy(matrix)
    for row in sorted_matrix:
        gap = len(row) // 2
        while gap > 0:
            for i in range(gap, len(row)):
                temp = row[i]
                j = i
                while j >= gap and row[j - gap] > temp:
                    row[j] = row[j - gap]
                    j -= gap
                row[j] = temp
            gap //= 2
    return sorted_matrix

# Função de Quick Sort
def quick_sort(matrix):
    def quicksort_helper(arr):
        if len(arr) <= 1:
            return arr
        pivot = arr[len(arr) // 2]
        left = [x for x in arr if x < pivot]
        middle = [x for x in arr if x == pivot]
        right = [x for x in arr if x > pivot]
        return quicksort_helper(left) + middle + quicksort_helper(right)

    sorted_matrix = copy.deepcopy(matrix)
    for i in range(len(sorted_matrix)):
        sorted_matrix[i] = quicksort_helper(sorted_matrix[i])
    return sorted_matrix

# Função de Heap Sort (Torneio)
import heapq
def heap_sort(matrix):
    sorted_matrix = copy.deepcopy(matrix)
    for i in range(len(sorted_matrix)):
        heapq.heapify(sorted_matrix[i])
        sorted_matrix[i] = [heapq.heappop(sorted_matrix[i]) for _ in range(len(sorted_matrix[i]))]
    return sorted_matrix

# Função para ordenação padrão do NumPy
def numpy_sort(matrix):
    return np.sort(matrix, axis=1)

# Teste e medição de tempo para cada algoritmo
algorithms = {
    "Selection Sort": selection_sort,
    "Insertion Sort": insertion_sort,
    "Bubble Sort": bubble_sort,
    "Shell Sort": shell_sort,
    "Quick Sort": quick_sort,
    "Heap Sort": heap_sort,
    "NumPy Sort": numpy_sort
}

# Dicionário para armazenar os tempos de execução
execution_times = {}

print("\nResultados das Ordenações:")
for name, sort_function in algorithms.items():
    start_time = time.time()
    sorted_matrix = sort_function(matrix)
    execution_time = time.time() - start_time
    execution_times[name] = execution_time

    print(f"\n{name}:")
    print(sorted_matrix)
    print(f"Tempo: {execution_time:.6f} segundos")

# Comparação de tempos
print("\n--- Comparação de Tempo de Execução ---")
for name, exec_time in execution_times.items():
    print(f"{name}: {exec_time:.6f} segundos")
```
